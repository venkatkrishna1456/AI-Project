```python
nb_closest_images = 10 # number of most similar images to retrieve
def retrieve_most_similar_products(given_img):

    print("***********************************************************************")
    print("original product:")
    directory= "style"
    directory=directory+'\\'+given_img

    original = load_img(directory, target_size=(150,150))
    plt.imshow(original)
    plt.show()

    print("**************************************************************************")
    print("most similar products:")

    closest_imgs = euc_distance_df_0[given_img].sort_values(ascending=False)[1:nb_closest_images+1].index
    closest_imgs_scores = euc_distance_df_0[given_img].sort_values(ascending=False)[1:nb_closest_images+1]

    for i in range(0,len(closest_imgs)):
        directory= "style"
        directory=directory+'\\'+closest_imgs[i]
        original = load_img(directory, target_size=(150,150))
        plt.imshow(original)
        plt.show()
        print("similarity score : ",closest_imgs_scores[i])
```
